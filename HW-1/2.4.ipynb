```python
from PIL import Image, ImageColor
img = Image.new('RGB', (800, 600))
```

```python
def generate(steps):
    data = [[1], [1, 1]]
    
    if steps == 1:
        return [[1]]     

    elif steps == 2:
        return data

    
    for step in range(2, steps):
        # start new data with 1
        new_data = [1]
        
        # generate other numbers
        for i in range(len(data[step-1]) - 1):
            new_data.append(data[step-1][i] + data[step-1][i+1])
        
        # finish data with another 1
        new_data.append(1)

        data.append(new_data)
        
    return data
```

```python
for step in generate(steps=7):
    print(step)
```

```python
def draw(steps):
    last_step_starting_x, y = 400, 100
    data = generate(steps)
    
    for step in range(steps):
        x = last_step_starting_x - 1
        for d in data[step]:
            if d%2 == 0:
                img.putpixel((x, y), ImageColor.getcolor('red', '1'))
            else:
                img.putpixel((x, y), ImageColor.getcolor('green', '1'))
            x += 2
        
        last_step_starting_x -= 1
        y += 1   
        
    img.save("2.4.jpg")
```

```python
draw(300)
```
