```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = Y = np.linspace(-1.5, 1.5, 500)
X, Y = np.meshgrid(X, Y)

# if distance of a point become greater than max_distance that point goes out
max_distance = 10
```

```python
def calculate_jump_out_step(x: np.ndarray, y: np.ndarray, c: "a + bj", steps: int) -> np.ndarray:
    '''
        get an array of x and an array of y and a 'c' and do some calculation for each point
        and return that each point at which step went out (its distance to the origin became
        greater than max_distance variable
        
        steps variable means the max steps to do calculations
    '''
    
    z = x + y*1j
    
    # at the initiation step all points stay in so z*0 == 0 means True for every points
    stay_in = np.where (z*0 == 0)
    
    # at the initiation step non of the points jump out so np.zeros means False for every points
    jump_out = np.zeros(z.shape)
    
    
    for s in range(steps):
        # do the function for staied in points
        z[stay_in] = z[stay_in]**2 + c
        
        # calculate which point still stay in after this step
        stay_in = np.where( np.abs(z) < max_distance )
        
        # if a point jump out its corresponding value in jump_out matrix will increase by 1
        jump_out += np.abs(z) > max_distance
        
    # after all calculation we do this because of our implementation of jump_out matrix
    jump_out_step = steps - jump_out
        
    return jump_out_step
```

```python
def draw(x: np.ndarray, y: np.ndarray, c: "a + bj", steps: int, cmap="tab20"):
    jump_out_step = calculate_jump_out_step(x, y, c, steps)
    
    plt.imshow(jump_out_step, origin="lower", cmap=cmap)
    
    plt.title(f"C = {c}")
    
    plt.colorbar()
```

```python
draw(X, Y, -1j, 50)
```

```python
draw(X, Y, -0.4-0.6j, 500)
```

```python
draw(X, Y, -0.6, 30)
```

```python
draw(X, Y, -0.12-0.75j, 50)
```

```python
draw(X, Y, -2, 100)
```

```python
draw(X, Y, -0.8+0.16j, 1500)
```

```python
draw(X, Y, -0.4+0.6j, 1000)
```
