```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from seaborn import set_theme; set_theme()
```

```python
def generate(N: int, plot=True):
    generated_numbers = np.zeros(N)
    n = 0
    while n < N:
        number = np.random.randint(10)
        if number == 4:
            generated_numbers[n] = np.random.randint(10)
            n += 1
        
    if plot == True:
        plt.figure()
        sn.displot(generated_numbers, kde=True)
        plt.xlabel("number")
        plt.ylabel("count")
        plt.title(f"N = {N}")
    
    return generated_numbers
```

```python
generate(100)
generate(1000)
generate(10000)
generate(100000);
```

```python
def calculate():
    max_N = int(input("max_N: "))
    std = np.zeros(max_N)
    
    for N in range(1, max_N):
        numbers = generate(N, plot=False)
        
        std_N = np.sqrt((numbers**2).mean() - numbers.mean()**2)
        
        std[N] = std_N/N
        
    N = np.arange(1, max_N)
    
    plt.plot(N, std[1:], label=r"$\frac{\sigma}{N}$")
    plt.plot(N, 1/np.sqrt(N), label=r"$\frac{1}{\sqrt{N}}$")
    
    plt.xlabel("N")
    plt.ylabel(r"$\frac{\sigma}{N}$")
    plt.title(f"max_N={max_N}")
    plt.legend()
```

```python
calculate()
```
