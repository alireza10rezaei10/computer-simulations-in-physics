```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
sn.set_theme()
```

```python
def generate(number_of_samples: int, N: int) -> np.ndarray:
    '''generate number_of_samples random samples (numbers) that each of them is sum of N other random numbers'''
    
    numbers = np.random.randint(10, size=number_of_samples*N).reshape([number_of_samples, N]).sum(axis=1)
    
    return numbers
```

```python
def draw():
    number_of_samples = int(input("number of samples: "))
    N = int(input("N: "))    
    
    numbers = generate(number_of_samples, N)
    
    sn.displot(numbers, kde=True)    
    
    plt.xlabel("number")
    plt.title(f"N = {N}")
    plt.show()
```

```python
draw()
```
