```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
sn.set_theme()
```

```python
def generate_gaussian_distribution(x1:np.ndarray, x2: np.ndarray, sigma: float) -> np.ndarray:
    rho = np.sqrt(-2*sigma**2*np.log(1-x1))
    theta = 2*np.pi*x2
    
    y1 = rho*np.sin(theta)
    y2 = rho*np.cos(theta)
    
    return y1, y2
```

```python
def draw():
    number_of_samples = int(input("number of samples: "))
    sigma = float(input("sigma: "))

    x1 = np.random.random(size=number_of_samples)
    x2 = np.random.random(size=number_of_samples)

    y1, y2 = generate_gaussian_distribution(x1, x2, sigma=1)
    
    sn.displot(x1, kde=True)
    plt.title(fr"x1, $\sigma={sigma}$")
    plt.xlabel("number")
    
    sn.displot(x2, kde=True)
    plt.title(fr"x2, $\sigma={sigma}$")
    plt.xlabel("number")
    
    sn.displot(y1, kde=True)
    plt.title(fr"y1, $\sigma={sigma}$")
    plt.xlabel("number")
    
    sn.displot(y2, kde=True)
    plt.title(fr"y2, $\sigma={sigma}$")
    plt.xlabel("number")
    
    plt.show()
```

```python
draw()
```
